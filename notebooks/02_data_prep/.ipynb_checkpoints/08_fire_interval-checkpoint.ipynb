```python
df = modelDataI=gpd.read_parquet("../../data/data_v07_landslide.parquet")
print('raw data -',df.shape)

# df=df.dropna()
# print('after nulls removed -',df.shape)

df.head()
```

```python
cols = [
    'year', # year of wildfire
    'stormstart', # date of storm that produced DF
]

df[cols]
```

```python
df['fire_interval'] = pd.to_datetime(df['stormstart']).dt.year - df['year']
df
```

```python
df['fire_interval'].value_counts(dropna=False)
```

```python
# encode nans'
df['fire_interval'] = df['fire_interval'].fillna(-9999)
```

```python
df['fire_interval'].value_counts(dropna=False)
```

```python
# there is small but not insignificant proportion where the rain event is before the fire
df[df['fire_interval'] < 0][['year', 'stormstart', 'response']]
```

```python
df.shape
```

```python
# dropping those cases
df = df[~(df['fire_interval'] < 0)]
df.shape
```
